```python
####### LIBRARIES
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score
```

```python
# MCS (sessions)
MCS= [0.15223, 0.14343, 0.13811, 0.13842, 0.1595, 0.1397, 0.1464, 0.1385, 0.1351, 0.1627, 0.1424, 0.1433, 0.1508, 0.1552, 0.1603, 0.14347, 0.14486, 0.14316]

# Gamma passing rate (sessions)
Gamma = [96.5, 96.0, 93.6, 95.9, 97.3, 96.2, 92.3, 96.1, 95.5, 98.2, 94.0, 95.3, 95.9, 93.1, 94.6, 94.7, 95.5, 98.1]
```

```python
####### MCS vs IG
def MCSGIA2(MCS, GI):    
    plt.figure(figsize=(8, 6))
    coef = np.polyfit(MCS, GI, 1)
    m = coef[0]
    b = coef[1]
    plt.scatter(
            MCS, GI,
            s=100,
            edgecolor='black', 
            )
    # Linear Fit
    x_fit = np.linspace(min(MCS), max(MCS), 100)
    y_fit = m * x_fit + b
    plt.plot(
        x_fit, y_fit,
        color='red',
        linestyle='--',
        linewidth=2,
    )
    # R^2
    y_pred = m*np.array(MCS)+b
    R2 = r2_score(GI,y_pred)
    
    plt.xlabel("MCS [a.u.]")
    plt.ylim(90, 100)
    plt.ylabel("Gamma passing rate [%]")
    plt.tight_layout()
    plt.show()
    print(f"y = {m:.2f}x + {b:.2f}")
    print(f"R² = {R2:.3f}")
```

```python
MCSGIA2(MCS,Gamma)
```
